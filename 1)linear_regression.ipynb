```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import os
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(0)  # Per riproducibilit√†
n_samples = 50
prezzi = np.random.uniform(5, 50, n_samples)  # Prezzi casuali tra 5 e 50
vendite = 100 - 2 * prezzi + np.random.normal(0, 10, n_samples)  # Vendite dipendenti dal prezzo con rumore aggiunto

# Creazione del DataFrame
df = pd.DataFrame({
    'Prezzo': prezzi,
    'Vendite': vendite
})


x = np.asarray(df.iloc[:,0]).reshape(-1,1)
y = np.asarray(df.iloc[:,1])

lr = LinearRegression()
lr.fit(x, y)
y_pred = lr.predict(x)
print(lr.intercept_)
print(lr.coef_)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Dati originali')
plt.plot(x, y_pred, color='red', label='Linear di regressione')
plt.xlabel('Prezzo')
plt.ylabel('Vendite')
plt.title('Regressione Lineare: Prezzo vs Vendite')
plt.legend()
plt.show()
```
