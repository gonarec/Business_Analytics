```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import os
import statsmodels.api as sm
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import metrics
```

```python
df=pd.read_csv('Dataset\Prostate_Cancer.csv')
x = df.iloc[:,[2,3,4,5,6,7,8,9]].values
y = df.iloc[:,1].values
df
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size = 0.2, random_state = 12345)

st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.transform(x_test)
```

```python
classifier= SVC(kernel='rbf',random_state=0)
classifier.fit(x_train,y_train)

classifier1= SVC(kernel='linear',random_state=0)
classifier1.fit(x_train,y_train)

classifier2= SVC(kernel='poly',random_state=0)
classifier2.fit(x_train,y_train)


y_pred = classifier.predict(x_test)
y_pred1 = classifier1.predict(x_test)
y_pred2 = classifier2.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)

# Stampa la matrice di confusione
df_cm = pd.DataFrame(cm, index=['True 0', 'True 1'], columns=['Predicted 0', 'Predicted 1'])
print('Accuracy:',metrics.accuracy_score(y_test,y_pred)*100,'%')
print(df_cm)
print()

cm1 = confusion_matrix(y_test, y_pred1)
print('\n')
# Stampa la matrice di confusione
df_cm1 = pd.DataFrame(cm1, index=['True 0', 'True 1'], columns=['Predicted 0', 'Predicted 1'])
print('Accuracy:',metrics.accuracy_score(y_test,y_pred1)*100,'%')
print(df_cm1)


cm2 = confusion_matrix(y_test, y_pred2)
print('\n')
# Stampa la matrice di confusione
df_cm2 = pd.DataFrame(cm2, index=['True 0', 'True 1'], columns=['Predicted 0', 'Predicted 1'])
print('Accuracy:',metrics.accuracy_score(y_test,y_pred2)*100,'%')
print(df_cm2)
```
