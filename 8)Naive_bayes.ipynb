```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
import seaborn as sns
import statsmodels.api as sm
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
df=pd.read_csv('Dataset/bill_authentication.csv')
x = df.iloc[:,[0,1,2,3]].values
y = df.iloc[:,4].values
df.head()
```

```python
x0=df.iloc[:,[0]].values
x1=df.iloc[:,[1]].values
x2=df.iloc[:,[2]].values
x3=df.iloc[:,[3]].values
```

```python
fig, axes = plt.subplots(2,2,figsize=(10, 6))

# Plot di tutti gli istogrammi sulla stessa figura
sns.histplot(x0, bins=30, kde=True, ax=axes[0,0], color='blue', label='x0')
axes[0,0].set_title
sns.histplot(x1, bins=30, kde=True, ax=axes[0,1], color='red', label='x1')
sns.histplot(x2, bins=30, kde=True, ax=axes[1,0], color='green', label='x2')
sns.histplot(x3, bins=30, kde=True, ax=axes[1,1], color='purple', label='x3')

# Aggiungere una legenda
plt.tight_layout()

# Mostrare il grafico
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size = 0.2, random_state = 12345)

st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.transform(x_test)
```

```python
Gau=GaussianNB()
Gau.fit(x_train,y_train)
y_pred=Gau.predict(x_test)
```

```python
print('Accuracy:',metrics.accuracy_score(y_test,y_pred)*100,'%')
```

```python
cm=confusion_matrix(y_test,y_pred)
df_cm = pd.DataFrame(cm, index=['True 0', 'True 1'], columns=['Predicted 0', 'Predicted 1'])
print(df_cm)
```
